```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
plt.figure(figsize=(5,5))
plt.plot([0, 0.25, 2], [0, 1, 0.5])
plt.show()
```

```python
plt.figure()
x = np.arange(-20, 21, 0.01)
plt.plot(x, x*x-3*x+2)
plt.show()
```

```python
plt.figure()
x = np.linspace(0,  10*np.pi, 100)

plt.figure(figsize=[10, 1])
plt.plot(x, np.sin(x))
plt.xlim = (-1, 34)
plt.show()
```

```python
#пример графика вообще со всем

x = np.linspace(0, 2 * np.pi, 100)

plt.figure(figsize=(10, 5))
plt.grid(color="red", linestyle='--', linewidth=0.5)
plt.plot(x, np.log(x), linewidth=2, color="orange", dashes=[8, 4], label=r'$\ln x$')
plt.plot(x, np.cos(x), linewidth=4, color="violet", dashes=[8, 4, 2, 4], label=r'$\cos x$')
plt.axis([0, 2 * np.pi, -1, 1])
plt.xticks(np.linspace(0, 2 * np.pi, 9),  # Где сделать отметки
           ('0',r'$\frac{1}{4}\pi$',r'$\frac{1}{2}\pi$',  # Как подписать
            r'$\frac{3}{4}\pi$',r'$\pi$',r'$\frac{5}{4}\pi$',
            r'$\frac{3}{2}\pi$',r'$\frac{7}{4}\pi$',r'$2\pi$'),
           fontsize=20)
plt.yticks(fontsize=12)
plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$y$', fontsize=20)
plt.title(r'$\sin x$, $\cos x$', fontsize=20)
plt.legend(fontsize=20, loc=1)
plt.show()
```
